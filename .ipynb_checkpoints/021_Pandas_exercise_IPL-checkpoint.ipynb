```python
data[data['city']=='Mumbai'].shape[0] # number of matches played in Mumbai
```

```python
# Creating a function to filter the data by city as an argument
```

```python
def match_city(place):
    mask1 = data['city']==place
    return data[mask1].shape[0]

match_city('Mumbai')
```
